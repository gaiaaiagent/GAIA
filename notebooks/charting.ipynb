```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import pandas as pd
import numpy as np

# Create the data
df = pd.DataFrame(zip(np.random.randn(20), np.random.randn(20)))

# Initialize Bokeh for notebook display
output_notebook()

# Create the figure
p = figure(width=700, height=400, title="Random Data Plot")

# Add the line
p.line(df.index, df[0], line_color='blue', line_width=2)

# Add some interactive tools
p.add_tools(bokeh.models.HoverTool())

# Set labels
p.xaxis.axis_label = 'Index'
p.yaxis.axis_label = 'Value'

# Show the plot
show(p)
```

```python
# Altair version
import altair as alt
import pandas as pd
import numpy as np
alt.renderers.enable('default')
# Create data
df = pd.DataFrame(zip(np.random.randn(20), np.random.randn(20)), columns=['a','b'])

# Create Altair chart
chart = alt.Chart(df.reset_index()).mark_line().encode(
    x='index:Q',
    y='0:Q'
).properties(
    width=700,
    height=400
)

chart
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Create data
df = pd.DataFrame(zip(np.random.randn(20), np.random.randn(20)))

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(df.index, df[0], '-b')  # blue line

# Add labels and title
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Random Data Plot')

# Add grid
plt.grid(True, linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```
